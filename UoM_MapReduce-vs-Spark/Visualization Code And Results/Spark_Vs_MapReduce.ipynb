````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

Creating a Dataframe Using the CSV File

````python
delay_df = pd.read_csv("MapReduceVsSpark.csv")
print(delay_df)
````




Getting the Comparison Graph for All 5 Attempts of Carrier Delay

````python
attempts = delay_df.iloc[0:5,1]
spark = delay_df.iloc[0:5,2]
hadoop = delay_df.iloc[0:5,3]
#spark_val = list(spark.values())

X_axis = np.arange(1,len(spark)+1)

plt.figure(figsize=(6, 6))
plt.bar(X_axis- 0.2, spark,0.4 ,label = 'Spark')
for index,value in enumerate(spark):
  plt.text(index+0.6,value+0.1,str(round(value,2)))
plt.bar(X_axis+ 0.2, hadoop,0.4 ,label = 'Hadoop')
for index,value in enumerate(hadoop):
  plt.text(index+1,value+0.1,str(round(value,2)))

plt.xticks(X_axis, attempts)
plt.xlim(0,6)
plt.ylim(0,12)
plt.xlabel("Iteration")
plt.ylabel("Runtime(s)")
plt.title("Year-wise Carrier Delay Query Runtime Comparison")
plt.grid(axis="y")
plt.legend(loc='upper left')
plt.show()

avg_time_carrier_spark = spark.mean()
avg_time_carrier_mapreduce=hadoop.mean()
print(avg_time_carrier_spark)
print(avg_time_carrier_mapreduce)
````
